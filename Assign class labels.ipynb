```python
import numpy as np
import pandas as pd
from collections import Counter
```

```python
df = pd.read_excel('preprocessed_mean.xlsx')
df1 = pd.read_excel('preprocessed_mode.xlsx')
df.head()
```

```python
#adding class label column
df["Class Label"] = np.nan
df1['Class Label'] = np.nan
df.head()
```

```python
#class label for df
for i in range(df.shape[0]):
  mghard = (df["MG"][i]*50.045)/20.04
  cahard = (df["CA"][i]*50.045)/12.15
  totalhard = (df["CA"][i]*2.497) + (df["MG"][i]*4.118)
  if df["PH"][i]>=6.5 and df["PH"][i]<=8.5 and mghard<200 and cahard<200 and totalhard<200 and df["CHLORIDE"][i]<250 and df["FLUORIDE"][i]<1.5 and df["NITRATE"][i]<20 and df["SULPHATE"][i]<400 :
    df["Class Label"][i]='A'
  elif df["PH"][i]>=6.5 and df["PH"][i]<=8.5 and df["FLUORIDE"][i]<1.5:
    df["Class Label"][i]='B' 
  elif df["PH"][i]>=6.0 and df["PH"][i]<=9.0 and df["CHLORIDE"][i]<600 and df["FLUORIDE"][i]<1.5 and df["NITRATE"][i]<50 and df["SULPHATE"][i]<400:
    df["Class Label"][i]='C'
  elif df["PH"][i]>=6.5 and df["PH"][i]<=8.5:
    df["Class Label"][i]='D'
  elif df["PH"][i]>=6.0 and df["PH"][i]<=8.5 and df["CHLORIDE"][i]<600 and df["EC"][i]<2250 and df["SULPHATE"][i]<1000 and df["SAR"][i]<26:
    df["Class Label"][i]='E'
  else:
    df["Class Label"][i]="Below E"
```
