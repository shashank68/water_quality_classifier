```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

```python
data = pd.read_excel('WATER QUALITY AGG DATA BY VENKI.xls', sheet_name = 'INTEGRATED DATA SET')
```

```python
data.head()
```

```python
data.shape
```

```python
sitename = data.loc[ :, ['SITENAME']]
sitename
```

```python
data1 = data.drop(['SITENAME'], axis=1)
data1
```

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values = np.NaN, strategy = 'mean')
imp1 = SimpleImputer(missing_values = np.NaN, strategy = 'most_frequent')
```

```python
imp.fit(data1)
imp1.fit(data1)
```

```python
y = imp.transform(data1)
y1 = imp1.transform(data1)
```

```python
col = ['ID', 'MG', 'PH', 'K', 'NITRATE', 'SULPHATE', 'EC', 'CA', 'NA', 'CARBONATE', 'BICARBONATE', 'CHLORIDE', 'FLUORIDE', 'SAR', 'RSC', 'YEAR']
```

```python
df = pd.DataFrame(data = y, index = [i for i in range(data.shape[0])], columns = col)
df.insert(1, 'SITENME', sitename)
df1 = pd.DataFrame(data = y1, index = [i for i in range(data.shape[0])], columns = col)
df1.insert(1, 'SITENAME', sitename)
```
